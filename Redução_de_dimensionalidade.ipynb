#Redução de Dimensionalidade em Imagens para Redes Neurais

##DESCRIÇÃO
Seguindo o exemplo do algoritmo de binarização apresentado em nossa última aula, realize a implementação em Python para transformar uma imagem colorida para níveis de cinza (0 a 255) e para binarizada (0 e 255), preto e branco.

```python
# Importando as bibliotecas necessárias
from PIL import Image
import matplotlib.pyplot as plt

# Função para converter a imagem colorida para tons de cinza (0 a 255)
def convert_to_grayscale(image_path):
    # Abre a imagem
    img = Image.open(image_path)

    # Converte a imagem para tons de cinza
    grayscale_img = img.convert('L')  # 'L' mode é para tons de cinza

    return grayscale_img

# Função para exibir a imagem em tons de cinza
def display_image(img, title):
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')  # Desliga os eixos
    plt.show()

# Caminho para a imagem (substitua pelo seu caminho ou faça o upload no Colab)
image_path = "/content/dog2.jpg"  # Substitua com o caminho da sua imagem

# Converte a imagem para tons de cinza
grayscale_image = convert_to_grayscale(image_path)

# Exibe a imagem em tons de cinza
display_image(grayscale_image, "Imagem em Tons de Cinza")
```

```python
# Função para binarizar a imagem
def binarize_image(grayscale_img, threshold=128):
    # Converte a imagem em um array de pixels
    img_array = grayscale_img.load()

    # Obtém as dimensões da imagem
    width, height = grayscale_img.size

    # Cria uma nova imagem em modo 'L' para armazenar os valores binarizados
    binary_img = Image.new('L', (width, height))
    binary_img_array = binary_img.load()

    # Itera sobre cada pixel da imagem em tons de cinza e aplica o limiar
    for y in range(height):
        for x in range(width):
            pixel_value = img_array[x, y]
            # Se o valor do pixel for maior que o limiar, torna-se branco (255), caso contrário, preto (0)
            binary_img_array[x, y] = 255 if pixel_value > threshold else 0

    return binary_img

# Função para exibir a imagem binarizada
def display_image(img, title):
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')  # Desliga os eixos
    plt.show()

# Binariza a imagem convertida para tons de cinza
binary_image = binarize_image(grayscale_image, threshold=128)

# Exibe a imagem binarizada
display_image(binary_image, "Imagem Binarizada (Preto e Branco)")
```
